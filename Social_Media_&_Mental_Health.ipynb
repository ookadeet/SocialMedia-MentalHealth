### Platforms Used

```python
print(df.Platforms.unique())
```

The Platforms column contains multiple answers per row. There is use in knowing which platforms, and how many in total are in use, so one-hot coding was done by creating new columns for each platform, assigning a 1 where it was in use, and 0 where it was not.

```python
# List of column names to be checked
column_names = ['Discord', 'Facebook', 'Instagram', 'Pinterest', 'Reddit', 'Snapchat', 'TikTok', 'Twitter', 'YouTube']

# Create new columns with 0 as the default value
for col_name in column_names:
    df[col_name] = df['Platforms'].str.contains(col_name, case=False).astype(int)
```

```python
#Double check the new columns work as expected
df.head()
```

```python
#Drop the Platforms column as no longer needed
df.drop('Platforms', axis=1, inplace=True)
```

```python
#Get a sum of the number of platforms used
df["Total Platforms"] = df.Discord + df.Facebook + df.Instagram + df.Pinterest + df.Reddit + df.Snapchat + df.TikTok + df.Twitter + df.YouTube
```

```python
#Reorder columns
desired_order = [
    'Age',
    'Gender',
    'Relationship Status',
    'Occupation',
    'Discord',
    'Facebook',
    'Instagram',
    'Pinterest',
    'Reddit',
    'Snapchat',
    'TikTok',
    'Twitter',
    'YouTube',
    'Total Platforms',
    'Hours Online',
    'ADHD1',
    'ADHD2',
    'ADHD3',
    'ADHD4',
    'Anx1',
    'Anx2',
    'SE1',
    'SE2',
    'SE3',
    'Dep1',
    'Dep2',
    'Dep3'
]

# Use reindex to rearrange the columns based on the desired_order
df = df.reindex(columns=desired_order)
```

### Hours Spent Online

```python
#Check answers to hours online
print(df['Hours Online'].unique())
```

The answers to how many hours are spent online was converted to a number to indicate scale, using the midpoint number. 0.5 was assigned to "Less than an hour" as it does not necessarily mean no time is spent online.

```python
#Use midpoint of "between x and y" as total hours number
df['Hours Online'] = df['Hours Online'].replace('Less than an Hour',0.5)
df['Hours Online'] = df['Hours Online'].replace('Between 1 and 2 hours',1.5)
df['Hours Online'] = df['Hours Online'].replace('Between 2 and 3 hours',2.5)
df['Hours Online'] = df['Hours Online'].replace('Between 3 and 4 hours',3.5)
df['Hours Online'] = df['Hours Online'].replace('Between 4 and 5 hours',4.5)
df['Hours Online'] = df['Hours Online'].replace('More than 5 hours',5.5)
df.info()
```

#Regression Modelling


## ADHD

```python
#Analysis on both independent variables
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
adhd_X = df[['Hours Online', 'Total Platforms']]
adhd_y = df['ADHD Total']

# Split the data
adhd_X_train, adhd_X_test, adhd_y_train, adhd_y_test = train_test_split(adhd_X, adhd_y, test_size=0.2, random_state=42)

# Add a constant to the X matrix
adhd_X_train_with_const = sm.add_constant(adhd_X_train)

# Fit the OLS model
adhd_model = sm.OLS(adhd_y_train, adhd_X_train_with_const).fit()

# Get p-values
adhd_p_values = adhd_model.pvalues

# Print results
for idx, col in enumerate(adhd_X_train.columns):
    if adhd_p_values[idx+1] < 0.05:
        print(f'{col}: Coefficient is statistically significant (p-value: {adhd_p_values[idx+1]})')
    else:
        print(f'{col}: Coefficient is not statistically significant (p-value: {adhd_p_values[idx+1]})')
```

```python
#Rerun without Total Platforms
adhd_X_hrs = df[['Hours Online']]

# Split the data
adhd_X_hrs_train, adhd_X_hrs_test, adhd_y_train, adhd_y_test = train_test_split(adhd_X_hrs, adhd_y, test_size=0.2, random_state=42)

# Add a constant to the X matrix
adhd_X_hrs_train_with_const = sm.add_constant(adhd_X_hrs_train)

# Fit the OLS model
adhd_hrs_model = sm.OLS(adhd_y_train, adhd_X_hrs_train_with_const).fit()

# Get p-values
adhd_hrs_p_values = adhd_hrs_model.pvalues

# Print results
for idx, col in enumerate(adhd_X_hrs_train.columns):
    if adhd_p_values[idx+1] < 0.05:
        print(f'{col}: Coefficient is statistically significant (p-value: {adhd_hrs_p_values[idx+1]})')
    else:
        print(f'{col}: Coefficient is not statistically significant (p-value: {adhd_hrs_p_values[idx+1]})')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Predict on the test set
adhd_X_hrs_test_with_const = sm.add_constant(adhd_X_hrs_test)
adhd_y_pred = adhd_hrs_model.predict(adhd_X_hrs_test_with_const)

# Calculate metrics
adhd_r_squared = r2_score(adhd_y_test, adhd_y_pred)
adhd_mae = mean_absolute_error(adhd_y_test, adhd_y_pred)
adhd_mse = mean_squared_error(adhd_y_test, adhd_y_pred)
adhd_rmse = np.sqrt(adhd_mse)

# Print metrics
print(f'Intercept: {adhd_hrs_model.params[0]}')
print(f'Coefficient for Hours Online: {adhd_hrs_model.params["Hours Online"]}')
print(f'R-squared: {adhd_r_squared}')
print(f'Mean Absolute Error (MAE): {adhd_mae}')
print(f'Mean Squared Error (MSE): {adhd_mse}')
print(f'Root Mean Squared Error (RMSE): {adhd_rmse}')
```

While Total Platforms used is not a statistically significant influencer of ADHD score, Hours Online are, with a 1.09 daily increase in hours online leading to an additional 1 point on the ADHD measure. When tested however, Hours Online explains only 20.4% of the variance seen in ADHD score and MAE and RMSE are high at 2.84 and 3.57 respectively.
